```python
%autoreload

# load libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for column in phy_df.columns:
    fig, ax = plt.subplots(figsize=(15,5))
    plt.scatter(phy_df[column], phy_df['score'])
    a, b = np.polyfit(phy_df[column], phy_df['score'],1)
    plt.plot(phy_df[column], a*phy_df[column]+b, linestyle='--',dashes=(3, 3), color='red')
    plt.grid(visible=None)
    plt.xlabel(column)
    plt.ylabel('score')
    plt.title(f'Correlation of {column} and the score')
    plt.savefig(f'{column}_score.png')
    plt.show()
    
```

```python
for column in phy_df.columns:
    fig, ax = plt.subplots(figsize=(15,5))
    plt.boxplot(phy_df[column])
    plt.grid(visible=None)
    plt.title(f'{column} boxplot')
    plt.savefig(f'{column}_box.png')
    plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import tree
from sklearn.model_selection import train_test_split
import graphviz

phy_used_features = phy_features[:] #copy the list
phy_used_features.remove('tempMean')
phy_used_features.remove('tempSD')
x_train, x_test, y_train, y_test = train_test_split(phy_df[phy_used_features], phy_df['score'], test_size=0.2, random_state=42)

linreg = LinearRegression().fit(StandardScaler().fit_transform(x_train), y_train)
treereg = DecisionTreeRegressor(random_state=42, min_impurity_decrease=0.05).fit(x_train, y_train)

print(f'mae of linreg on train data {mean_absolute_error(linreg.predict(StandardScaler().fit_transform(x_train)), y_train)} score {r2_score(y_train, linreg.predict(StandardScaler().fit_transform(x_train)))}')
print(f'mae of treereg on train data {mean_absolute_error(treereg.predict(x_train), y_train)} score {r2_score(y_train, treereg.predict(x_train))}')
print(f'mae of linreg on test data{mean_absolute_error(linreg.predict(StandardScaler().fit_transform(x_test)), y_test)} score {r2_score(y_test, linreg.predict(StandardScaler().fit_transform(x_test)))}')
print(f'mae of treereg on test data{mean_absolute_error(treereg.predict(x_test), y_test)} score {r2_score(y_test, treereg.predict(x_test),)}')
print('mean absolute error of regression tree on all features is: 0.6657295008304702')
print(f'mean absolute error of just taking mean value is: {mean_absolute_error(np.full(y_test.shape, y_test.mean()) ,y_test)} score {r2_score(y_test, np.full(y_test.shape, y_test.mean()))}')

print('Linreg coefs: ')
for col, coef in zip(phy_df.columns[:-1], linreg.coef_):
    print("\t", col, coef)

dot_data = tree.export_graphviz(treereg, out_file=None, feature_names=phy_used_features, filled=True)
graph = graphviz.Source(dot_data, format="png")
graph.render("regression_tree")
graph
```
